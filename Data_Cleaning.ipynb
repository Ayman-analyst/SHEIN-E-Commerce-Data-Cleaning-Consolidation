# **Reading Files**

```python
df1  = pd.read_csv("e_commerce_data/us-shein-appliances-3987.csv")
df2  = pd.read_csv("e_commerce_data/us-shein-automotive-4110.csv")
df3  = pd.read_csv("e_commerce_data/us-shein-baby_and_maternity-4433.csv")
df4  = pd.read_csv("e_commerce_data/us-shein-bags_and_luggage-4299.csv")
df5  = pd.read_csv("e_commerce_data/us-shein-home_and_kitchen-3719.csv")
df6  = pd.read_csv("e_commerce_data/us-shein-home_textile-3883.csv")
df7  = pd.read_csv("e_commerce_data/us-shein-kids-4314.csv")
df8  = pd.read_csv("e_commerce_data/us-shein-office_and_school_supplies-4233.csv")
df9  = pd.read_csv("e_commerce_data/us-shein-tools_and_home_improvement-3903.csv")
df10 = pd.read_csv("e_commerce_data/us-shein-underwear_and_sleepwear-4019.csv")

dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10]
```

```python
cols =  [
    "Selling Proposition",
    "Price",
    "Discount",
    "Rank Title",
    "Rank Subcategory",
    "Product Title",
    "Color Count"
]


for i,df in enumerate(dataframes):
    print(f"DF{i+1} Columns Unique Values: ")
    print()
    for col in cols:
        if col in df.columns:
            unique_values = df[col].unique()
            print(f"Unique Values in {col} Column : {unique_values}")
            print()
        else:
            print(f"Unique Values in {col} Column : No Column")

    print()
    print()
```

```python
for df in dataframes:
    df["Price"]= df["Price"].str.replace("$","")
df1.head()
```

```python
for df in dataframes:
    if "Rank Subcategory" in df.columns:
        df["Rank Subcategory"] = df["Rank Subcategory"].str.replace("in ", "")

df1.head()
```

```python
for df in dataframes:
    df["Discount"] = df["Discount"].str.replace("%", "")
df1.head()
```

```python
for df in dataframes:
    # Convert all values to string first
    df["Selling Proposition"] = df["Selling Proposition"].astype(str).fillna("0")

    # Remove unwanted characters
    df["Selling Proposition"] = df["Selling Proposition"] \
        .str.replace("+", "", regex=False) \
        .str.replace("sold recently", "", regex=False)

    # Convert "k" values to numbers and multiply by 1000
    df["Selling Proposition"] = df["Selling Proposition"].apply(
        lambda x: float(x.replace("k", "")) * 1000 if "k" in x else float(x)
    )

    # Ensure numeric type
    df["Selling Proposition"] = pd.to_numeric(df["Selling Proposition"], errors="coerce").fillna(0)

df1.head()
```

```python
for i, df in enumerate(dataframes):
    print(f"DF{i+1} Columns Types")
    print(df.dtypes)
    print()
```

```python
for i, df in enumerate(dataframes):
    zz = df["Price"].isna().sum()
    print(f"DF {i + 1} has {zz} missing values in Price Column")
```

```python
for df in dataframes:
    # Fix "Color Count" only if it exists
    if "Color Count" in df.columns:
        df["Color Count"] = df["Color Count"].fillna(0).astype(int)

    # Ensure "Price" column has no commas and convert to float
    df["Price"] = df["Price"].fillna("").astype(str).str.replace(",", "", regex=True)
    df["Price"] = df["Price"].replace("", "0").astype(float)  # Handle empty values

    # Fix "Discount" column: fill NaN and convert to int
    df["Discount"] = df["Discount"].fillna(0).astype(int)
```
